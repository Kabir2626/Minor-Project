```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create a dummy expenses.csv file for demonstration
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Month_Number': [1, 2, 3, 4, 5, 6],
    'Salary': [5000, 5000, 5000, 5000, 5000, 5000],
    'Electricity': [100, 120, 110, 130, 115, 125],
    'Internet': [50, 50, 50, 50, 50, 50],
    'Transport': [80, 90, 85, 95, 88, 92],
    'Raw Material': [1000, 1100, 1050, 1150, 1080, 1120],
    'Misc': [200, 210, 205, 215, 208, 212]
}
df_dummy = pd.DataFrame(data)
df_dummy['Total'] = df_dummy[['Electricity', 'Internet', 'Transport', 'Raw Material', 'Misc']].sum(axis=1)
df_dummy.to_csv('expenses.csv', index=False)

# Load expense data
df = pd.read_csv("expenses.csv")

# Show totals
print("\n--- TOTAL EXPENSES BY MONTH --")
print(df[["Month", "Total"]])
# =世##：MACHINE LEARNING PART ：##=

# Using Month_Number to predict Total
x = df[["Month_Number"]] # Input (independent variable)
y = df["Total"]  # Output (dependent variable)

# Create regression model
model = LinearRegression()
model.fit(x, y)

# Predict next month expense
next_month = [[df["Month_Number"].max() + 1]]
predicted_expense = model.predict(next_month)

print("\nPredicted total expense for next month: ")
print("$", round(predicted_expense[0], 2))

#  VISUALIZATION  #
plt.scatter(x, y)
plt.plot(x, model.predict(x))
plt.title("Expense Prediction Using Linear Regression")
plt.xlabel("Month Number")
plt.ylabel("Total Expense")
plt.show()
# ==***== EXPENSE CATEGORY ANALYSIS =***=
category_totals = df[["Salary", "Electricity", "Internet", "Transport", "Raw Material", "Misc"]].sum()
max_category = category_totals.idxmax()
print("InHighest Spending Category:")
print(max_category, "=", category_totals[max_category])
```
